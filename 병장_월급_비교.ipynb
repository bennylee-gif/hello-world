```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 만들기 (군인 월급 데이터)
data = {
    'Year': [2010, 2012, 2014, 2016, 2018, 2020, 2022, 2024, 2025],
    'Salary': [97500, 108000, 149000, 197000, 405700, 540900, 676100, 1250000, 1500000]
}

# 데이터를 표(DataFrame) 형태로 변환
df = pd.DataFrame(data)

# 2. 그래프 그리기
plt.figure(figsize=(10, 6)) # 그래프 크기 설정
plt.plot(df['Year'], df['Salary'], marker='o', color='green', linestyle='-')
plt.title('Korea Soldier Salary Trend (2010-2025)') # 제목
plt.xlabel('Year') # 가로축 이름
plt.ylabel('Salary (KRW)') # 세로축 이름
plt.grid(True) # 격자 무늬 추가

# 그래프 보여주기
plt.show()

# 3. 결과 텍스트로 확인하기
print("데이터 분석 결과:")
print(df)
```

```python
# 치킨 1마리 가격 (대략 2만원으로 가정)
chicken_price = 20000

# 'Chicken_Count'라는 새로운 열(Column) 만들기
df['Chicken_Count'] = df['Salary'] / chicken_price

# 소수점 첫째 자리까지만 예쁘게 보여주기
pd.options.display.float_format = '{:.1f}'.format

print("=== 연도별 월급으로 사먹을 수 있는 치킨 마릿수 ===")
print(df[['Year', 'Chicken_Count']])
```
